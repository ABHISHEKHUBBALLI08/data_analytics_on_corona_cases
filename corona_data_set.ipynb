```python
da.isnull().sum()
```

```python
from sklearn.impute  import SimpleImputer
imputer=SimpleImputer(strategy='constant')
da.continent= imputer.fit_transform(da['continent'].values.reshape(-1,1))
da.excess_mortality_cumulative_absolute= imputer.fit_transform(da['excess_mortality_cumulative_absolute'].values.reshape(-1,1))
da.excess_mortality_cumulative= imputer.fit_transform(da['excess_mortality_cumulative'].values.reshape(-1,1))
da.excess_mortality= imputer.fit_transform(da['excess_mortality'].values.reshape(-1,1))
da.excess_mortality_cumulative_per_million= imputer.fit_transform(da['excess_mortality_cumulative_per_million'].values.reshape(-1,1))
```

```python
from sklearn.impute  import SimpleImputer
imputer=SimpleImputer(strategy='median')
da.total_cases= imputer.fit_transform(da['total_cases'].values.reshape(-1,1))
```

```python
da.isnull().sum()
```

```python
ax = sn.countplot(x = 'continent', data = da)
for bars in ax.containers:ax.bar_label(bars)
```

```python
Sales_gen = da.groupby(['continent'], as_index=False)['total_cases'].sum().sort_values(by='total_cases', ascending=False)
ax = sn.barplot(x= 'continent', y = 'total_cases', data = Sales_gen)
```

```python
Sales_state = da.groupby(['continent'], as_index=False)['population'].sum().sort_values(by = 'population', ascending=False).head(10)
sn.set(rc={'figure.figsize':(20,5)})
sn.barplot( x = 'continent', y = 'population', data = Sales_state)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
data.Amount = imputer.fit_transform(data['Amount'].values.reshape(-1,1))
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
ax = sns.countplot(x = 'State', data = data)
for bars in ax.containers:ax.bar_label(bars)
```

```python
Sales_gen = data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ax = sns.barplot(x= 'Gender', y = 'Amount', data = Sales_gen)
```

```python
ax = sns.countplot(x = 'Age Group', data = data, hue='Gender')
for bars in ax.containers:ax.bar_label(bars)
```

```python
Sales_age = data.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = Sales_age)
```

```python
Sales_state = data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot( x = 'State', y = 'Orders', data = Sales_state)
```
